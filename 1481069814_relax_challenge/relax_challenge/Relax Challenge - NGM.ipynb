### Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import itertools
import warnings
from datetime import datetime
from sklearn.model_selection import train_test_split, cross_validate, RandomizedSearchCV
from sklearn.pipeline import make_pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.dummy import DummyClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import f1_score
from sklearn.metrics import classification_report, confusion_matrix, ConfusionMatrixDisplay, f1_score, roc_auc_score,  roc_curve
```

### Read In CSV Data

```python
users = pd.read_csv('takehome_users.csv', encoding='latin-1')
usage = pd.read_csv('takehome_user_engagement.csv')
```

### View and Fix Datatypes

```python
users.dtypes
```

```python
users['object_id'] = users['object_id'].astype(object)

users['creation_time'] = pd.to_datetime(users.creation_time)

def convert_timestamp_to_date(timestamp):
    if pd.notna(timestamp):
        return datetime.utcfromtimestamp(timestamp).strftime('%Y-%m-%d %H:%M:%S')
    else:
        return None
    
users['last_session_creation_time'] = users['last_session_creation_time'].apply(convert_timestamp_to_date)
users['last_session_creation_time'] = pd.to_datetime(users['last_session_creation_time'])

users['org_id'] = users['org_id'].astype(object)

def cast_to_int(value):
    try:
        return int(value)
    except:
        return -99

users['invited_by_user_id'] = users['invited_by_user_id'].apply(cast_to_int).astype(object)
```

### View Usage Data

```python
usage.head()
```

### View and Fix Datatypes

```python
usage.dtypes
```

```python
usage['time_stamp'] = pd.to_datetime(usage.time_stamp)
```

### Get Dataframe of Unique Days that Users Logged In

```python
print(usage.visited.drop_duplicates())
usage['login_day'] = pd.to_datetime(usage.time_stamp.dt.date)
daily_usage = usage[['login_day', 'user_id']].drop_duplicates()
daily_usage = daily_usage.sort_values(by=['user_id', 'login_day'], ascending=[True, True])
```

### Create "adopted" Variable (Logged In Three or More Times In Any Seven Day Period)

```python
adopted = pd.DataFrame(daily_usage.groupby('user_id')['login_day'].diff(periods=2).dt.days)
adopted = adopted.rename({'login_day': 'num_days_three_logins'}, axis=1)
adopted = pd.merge(daily_usage, adopted, left_index=True, right_index=True)[['user_id', 'num_days_three_logins']]
adopted.dropna(inplace=True)
adopted = adopted.groupby(['user_id'])['num_days_three_logins'].min().reset_index()
adopted['adopted'] = [1 if x <= 7 else 0 for x in adopted.num_days_three_logins]
adopted = adopted[['user_id', 'adopted']]
```

### Merge "adopted" Users With Users Dataframe

```python
adopted_users = pd.merge(users, adopted, left_on='object_id', right_on='user_id', how='left').drop(['user_id'], axis=1)
adopted_users['adopted'] = adopted_users.adopted.fillna(0)
```

### Feature Engineering

```python
adopted_users['days_since_created'] = (adopted_users.last_session_creation_time.max() - adopted_users.creation_time).dt.days
adopted_users['days_since_last_session'] = (adopted_users.last_session_creation_time.max() - adopted_users.last_session_creation_time).dt.days.fillna(-99).astype('int64')
adopted_users = adopted_users.drop(['creation_time', 'last_session_creation_time'], axis=1)
```

```python
adopted_users['email_domain'] = adopted_users.email.str.split('@').str[1].str.split('.').str[0]
adopted_users['invited'] = [1 if x != -99 else 0 for x in adopted_users.invited_by_user_id]
```

```python
adopted_users = adopted_users[['object_id', 'email_domain', 'creation_source', 'org_id', 'opted_in_to_mailing_list', 
                               'enabled_for_marketing_drip', 'invited', 'days_since_created','days_since_last_session',
                               'adopted']]
```

```python
adopted_users.head()
```

```python
email_domain_counts = pd.DataFrame(adopted_users.email_domain.value_counts())
email_domain_counts.loc[email_domain_counts['count'] < 1000, 'email_domain_clean'] = 'other'
email_domain_counts.loc[email_domain_counts['count'] >= 1000, 'email_domain_clean'] = email_domain_counts.loc[email_domain_counts['count'] >= 1000].index
adopted_users = pd.merge(adopted_users, email_domain_counts['email_domain_clean'], left_on='email_domain', right_index=True, how='left').drop(['email_domain'], axis=1)
adopted_users = adopted_users.rename({'email_domain_clean': 'email_domain'}, axis=1)
```

```python
org_id_counts = pd.DataFrame(adopted_users.org_id.value_counts())
adopted_users = pd.merge(adopted_users, org_id_counts['count'], left_on='org_id', right_index=True, how='left').drop(['org_id'], axis=1)
adopted_users = adopted_users.rename({'count': 'size_of_org'}, axis=1)
```

### Model Training

#### Split data into training and testing splits

```python
X_train, X_test, y_train, y_test = train_test_split(adopted_users.drop(columns=['adopted', 'object_id']), adopted_users.adopted, test_size=0.3, 
                                                    random_state=47)
```

#### Dummy classifier

```python
dummy = DummyClassifier(strategy='constant', constant=1)
dummy.fit(X_train, y_train)
y_pred = dummy.predict(X_test)
f1_score(y_test, y_pred)
```

#### Logistic Regression

```python
pipe_log_reg = make_pipeline(
    SimpleImputer(), 
    StandardScaler(),
    LogisticRegression()
)
```

```python
grid_params = {
        'simpleimputer__strategy': ['mean', 'median'],
        'standardscaler': [StandardScaler(), MinMaxScaler(), None],
        'simpleimputer__strategy': ['mean', 'median'],
        'logisticregression__C': [0.01, 0.1, 1, 10, 100],
        'logisticregression__fit_intercept': [True, False],
        'logisticregression__l1_ratio': [0, .1, .25, .5, .75, 1],
        'logisticregression__penalty': ['elasticnet', None],
        'logisticregression__solver': ['lbfgs', 'liblinear', 'newton-cg', 'newton-cholesky', 'sag', 'saga']
}
```

```python
warnings.filterwarnings("ignore")
log_reg_randomized_cv = RandomizedSearchCV(pipe_log_reg, param_distributions=grid_params, cv=5, n_jobs=-1, 
                                           n_iter = 60, scoring='f1')
log_reg_randomized_cv.fit(X_train, y_train)
print(log_reg_randomized_cv.best_params_)
print(f'Best Score = {log_reg_randomized_cv.best_score_}')
```

#### Adaboost - Decision Tree Estimator 

```python
decision_tree = DecisionTreeClassifier()

pipe_adaboost_decision_tree = make_pipeline(
    SimpleImputer(), 
    StandardScaler(),
    AdaBoostClassifier()
)
```

```python
grid_params = {
    'adaboostclassifier__estimator': [decision_tree],
    'simpleimputer__strategy': ['mean', 'median'],
    'standardscaler': [StandardScaler(), MinMaxScaler(), None],
    'adaboostclassifier__algorithm': ['SAMME', 'SAMME.R'],
    'adaboostclassifier__learning_rate': [.01, .25, .5, .75, 1],
    'adaboostclassifier__n_estimators': [100, 200, 300, 400, 500],
    'adaboostclassifier__estimator__ccp_alpha': [.001, .01, .1, .25],
    'adaboostclassifier__estimator__criterion': ['gini', 'entropy', 'log_loss'],
    'adaboostclassifier__estimator__max_depth': [1, 2, 3],
    'adaboostclassifier__estimator__max_features': [None, 'auto', 'sqrt', 'log2'], 
    'adaboostclassifier__estimator__min_impurity_decrease': [.0001, .001, .01, .1],
    'adaboostclassifier__estimator__min_samples_leaf': [1, 2, 3, 4, 5],
    'adaboostclassifier__estimator__min_samples_split': [2, 3, 4, 5],
    'adaboostclassifier__estimator__splitter': ['best', 'random']
}
```

```python
adaboost_decision_tree_randomized_cv = RandomizedSearchCV(pipe_adaboost_decision_tree, param_distributions=grid_params,
                                                    cv=5, n_jobs=-1, n_iter = 60, scoring='f1')
adaboost_decision_tree_randomized_cv.fit(X_train, y_train)
print(adaboost_decision_tree_randomized_cv.best_params_)
print(f'Best Score = {adaboost_decision_tree_randomized_cv.best_score_}')
```

### Final Model Selection

```python
def f1_score_output(model, X_train, X_test, y_train, y_test):
    f1 = cross_validate(model.best_estimator_, X_train, y_train, scoring='f1', cv=5, n_jobs=-1)
    f1_mean = np.mean(f1['test_score'])
    f1_std = np.std(f1['test_score'])
    print('Cross-Validation (Training Data)')
    print(f1_mean, f1_std)
    print('Test Score')
    print(f1_score(y_test, model.best_estimator_.predict(X_test)))
```

#### Logistic Regression

```python
f1_score_output(log_reg_randomized_cv, X_train, X_test, y_train, y_test)
```

#### Adaboost - Decision Tree Estimator

```python
f1_score_output(adaboost_decision_tree_randomized_cv, X_train, X_test, y_train, y_test)
```

### Model Evaluation

#### Classification Report

```python
y_predict = adaboost_decision_tree_randomized_cv.best_estimator_.predict(X_test)
classification_report_dict = classification_report(y_test, y_predict, output_dict=True)
classification_report_df = pd.DataFrame(dict(itertools.islice(classification_report_dict.items(), 2))).T
classification_report_df = classification_report_df[['precision', 'recall', 'f1-score']]
long_df = classification_report_df.reset_index().melt(id_vars='index')
long_df.columns = ['adopted', 'Metric', 'Value']
long_df['adopted'] = np.where(long_df['adopted']=='1.0', 'active', 'not active')
```

```python
plt.figure(figsize=(10, 6))

bar_plot = sns.barplot(x='adopted', y='Value', hue='Metric', data=long_df)

for bar in bar_plot.patches:
    bar_plot.annotate(format(bar.get_height(), '.2f'),
                      (bar.get_x() + bar.get_width() / 2,
                       bar.get_height()), ha='center', va='center',
                       fontsize='large', xytext=(0, 8),
                       textcoords='offset points', fontweight='bold')

plt.title('Classification Report by adopted \n (Test Set)', fontsize=10, fontweight='bold')
plt.xlabel('adopted')
plt.ylabel('Value')
plt.legend(loc='lower right')
plt.show()
```

#### ROC curve

```python
y_pred_proba = adaboost_decision_tree_randomized_cv.best_estimator_.predict_proba(X_test)

fpr, tpr, thresholds = roc_curve(y_test, y_pred_proba[:, 1])

auc = roc_auc_score(y_test, y_pred_proba[:, 1])

plt.plot(fpr, tpr, label=f'AUC = {auc:.2f}')
plt.plot([0, 1], [0, 1], linestyle='--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.legend(loc='lower right')
plt.show()
```

Looking at the exploratory data analysis, the most notable differences between groups seemed to be that adopted users tended to have created their accounts sooner, have fewer days since their last session, and come from smaller organizations. This was backed up in the feature importances of our final model (Adaboost with Decision Tree classifer) where days_since_last_session, days_since_created, and size_of_org were the most important features. While the logistic regression model didn't perform much better than the dummy classifier, this model performed significantly better and was able to very accurately classify users as "adopted" or not. That being said, the first two of these variables might not be very informative when used together since users who created their accounts long ago and are still using the platform recently are very likely to have adopted it. Future models may want to exclude one of these variables to see if other features can help inform business decision making.
