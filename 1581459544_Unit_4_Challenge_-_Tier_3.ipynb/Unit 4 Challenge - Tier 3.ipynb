#### 1.2.  Loading the data
Your data comes from the [London Datastore](https://data.london.gov.uk/): a free, open-source data-sharing portal for London-oriented datasets. 

```python
# First, make a variable called url_LondonHousePrices, and assign it the following link, enclosed in quotation-marks as a string:
# https://data.london.gov.uk/download/uk-house-price-index/70ac0766-8902-4eb5-aab5-01951aaed773/UK%20House%20price%20index.xls

url_LondonHousePrices = "https://data.london.gov.uk/download/uk-house-price-index/70ac0766-8902-4eb5-aab5-01951aaed773/UK%20House%20price%20index.xls"

# The dataset we're interested in contains the Average prices of the houses, and is actually on a particular sheet of the Excel file. 
# As a result, we need to specify the sheet name in the read_excel() method.
# Put this data into a variable called properties.  
properties = pd.read_excel(url_LondonHousePrices, sheet_name='Average price', index_col= None)
```

**2.2. Cleaning the data**

You might find you need to transpose your dataframe, check out what its row indexes are, and reset the index. You  also might find you need to assign the values of the first row to your column headings  . (Hint: recall the .columns feature of DataFrames, as well as the iloc[] method).

Don't be afraid to use StackOverflow for help  with this.

```python
# Transpose the properties dataframe
properties_trans = properties.transpose().reset_index()
# Set the column names equal to the first row of data
properties_trans.columns = properties_trans.iloc[0]
# Delete the first row of data (now the column names)
properties_trans.drop(properties_trans.index[0], inplace=True)
# View the first five rows of the properties_trans dataframe
properties_trans.head()
```

**2.3. Cleaning the data (part 2)**

You might we have to **rename** a couple columns. How do you do this? The clue's pretty bold...

```python
# Rename the first two columns (London Borough and ID)
properties_trans.rename(columns={properties_trans.columns[0]: 'London Borough',
                        properties_trans.columns[1]: 'ID'}, inplace=True)
# View the first five rows
properties_trans.head()
```

**2.4.Transforming the data**

Remember what Wes McKinney said about tidy data? 

You might need to **melt** your DataFrame here. 

```python
# Melt the properties_trans dataframe so that there is one Month column and one Average Price column.
# Call the new dataframe properties_trans_melted
properties_trans_melted = properties_trans.melt(id_vars=['London Borough', 'ID'], var_name='Month', value_name='Average Price')
```

Remember to make sure your column data types are all correct. Average prices, for example, should be floating point numbers... 

```python
# Print the column names and the datatypes
print(properties_trans_melted.dtypes)
# Change the type of the London Borough and ID columns to string and change the type of the Average Price column to float
properties_trans_melted=properties_trans_melted.astype({'London Borough': 'string',
                               'ID': 'string',
                               'Average Price': 'float'})
# Print the column names and the datatypes
print(properties_trans_melted.dtypes)
```

**2.5. Cleaning the data (part 3)**

Do we have an equal number of observations in the ID, Average Price, Month, and London Borough columns? Remember that there are only 32 London Boroughs. How many entries do you have in that column? 

Check out the contents of the London Borough column, and if you find null values, get rid of them however you see fit. 

```python
# Print the number of non-null records in each column
print(f'Number of non-null records:\n{properties_trans_melted.count()}\n')
# Isolate the rows that are missing a value
missing_values_ID = properties_trans_melted.loc[properties_trans_melted['ID'].isnull()]
missing_values_Avg_Price = properties_trans_melted.loc[properties_trans_melted['Average Price'].isnull()]
missing_values = properties_trans_melted.loc[properties_trans_melted['ID'].isnull() |
                                            properties_trans_melted['Average Price'].isnull()]
# Confirm that all rows missing a value in the ID column are missing a value in the Average Price column
print(f'All rows missing an ID are missing an Average Price: {missing_values_ID.index.equals(missing_values_Avg_Price.index)}\n')
# Find all the London Boroughs that have null records
print(f'London Boroughs in rows missing a value:\n{missing_values["London Borough"].unique()}\n')
# Delete the null records since the London Boroughs are unnamed in those rows
properties_trans_melted = properties_trans_melted.drop(missing_values.index).reset_index(drop=True)
# Find the number of London Boroughs in the other rows
print(f'Number of London Boroughs in usable rows: {len(properties_trans_melted["London Borough"].unique())}')
# Look at the boroughs that are in the usable rows
print(f'London Boroughs in usable rows:\n{sorted(properties_trans_melted["London Borough"].unique())}\n')
# Create a list of the 32 London Boroughs
london_boroughs = ['Barking & Dagenham', 'Barnet', 'Bexley', 'Brent', 'Bromley', 'Camden','Croydon', 'Ealing', 'Enfield', 
                   'Greenwich', 'Hackney', 'Hammersmith & Fulham', 'Haringey', 'Harrow', 'Havering', 'Hillingdon','Hounslow', 
                   'Islington', 'Kensington & Chelsea', 'Kingston upon Thames','Lambeth','Lewisham', 'Merton', 'Newham', 
                   'Redbridge','Richmond upon Thames','Southwark', 'Sutton','Tower Hamlets', 'Waltham Forest', 'Wandsworth', 
                   'Westminster']
# Filter properties_trans_melted for just the rows that are for the 32 London Boroughs
properties_trans_melted = properties_trans_melted[properties_trans_melted['London Borough'].isin(london_boroughs)].reset_index(drop=True)
# Confirm that there are 32 boroughs in the dataset
print(f'Final number of London Boroughs in dataset: {len(properties_trans_melted["London Borough"].unique())}')
```

**2.6. Visualizing the data**

To visualize the data, why not subset on a particular London Borough? Maybe do a line plot of Month against Average Price?

```python
# Create a 4 X 8 grid of plots (one for each borough)
fig, axs = plt.subplots(8, 4, figsize=(32, 32))
# Loop through the list of boroughs
for i, borough in enumerate(london_boroughs):
    # Isolate the rows for the current borough in the loop
    data = properties_trans_melted.loc[properties_trans_melted['London Borough']==borough]
    # Find which plot corresponds to this iteration
    ax = axs[i // 4, i % 4]
    # Plot Average Price by Month for the borough
    ax.plot(data['Month'], data['Average Price'])
    # Set the title equal to the borough name
    ax.set_title(borough)
    # Adjust the y-axis scale
    ax.set_ylim(0, 2000000)
# Show the plots
plt.show()
```

To limit the number of data points you have, you might want to extract the year from every month value your *Month* column. 

To this end, you *could* apply a ***lambda function***. Your logic could work as follows:

```python
# Create a new column (Year) from the Month column
properties_trans_melted['Year'] = properties_trans_melted['Month'].apply(lambda x: x.year)
# Find the average price for each Borough, ID, Year combination
properties_trans_melted_by_year = properties_trans_melted.groupby(['London Borough', 'ID', 'Year']).agg({'Average Price': 'mean'}).reset_index()
# Plot Average Price by Year for each borough
fig, axs = plt.subplots(8, 4, figsize=(32, 32))
for i, borough in enumerate(london_boroughs):
    data = properties_trans_melted_by_year.loc[properties_trans_melted_by_year['London Borough']==borough]
    ax = axs[i // 4, i % 4]
    ax.plot(data['Year'], data['Average Price'])
    ax.set_title(borough)
    ax.set_ylim(0, 2000000)
plt.show()
```

**3. Modeling**

Consider creating a function that will calculate a ratio of house prices, comparing the price of a house in 2018 to the price in 1998.

Consider calling this function create_price_ratio.

You'd want this function to:
1. Take a filter of dfg, specifically where this filter constrains the London_Borough, as an argument. For example, one admissible argument should be: dfg[dfg['London_Borough']=='Camden'].
2. Get the Average Price for that Borough, for the years 1998 and 2018.
4. Calculate the ratio of the Average Price for 1998 divided by the Average Price for 2018.
5. Return that ratio.

Once you've written this function, you ultimately want to use it to iterate through all the unique London_Boroughs and work out the ratio capturing the difference of house prices between 1998 and 2018.

Bear in mind: you don't have to write a function like this if you don't want to. If you can solve the brief otherwise, then great! 

***Hint***: This section should test the skills you acquired in:
- Python Data Science Toolbox - Part One, all modules

```python
# Suppress scientific notation
pd.options.display.float_format = '{:.10f}'.format

# Function that takes a borough name and returns a dataframe of the borough name, its average housing price in 1998, 
# average housing price in 2018, and the ratio of the average housing price in 2018 relative to 1998
# (Higher values indicate that housing prices have increased by a greater %)
def create_price_ratio(borough):
    # Isolate years for the borough passed as an argument
    data = properties_trans_melted_by_year.loc[properties_trans_melted_by_year['London Borough']==borough]
    # Isolate row for 1998
    avg_borough_price_1998 = data.loc[data['Year']==1998, 'Average Price'].values
    # Isolate row for 2018
    avg_borough_price_2018 = data.loc[data['Year']==2018, 'Average Price'].values
    # Calculate price ratio
    price_ratio = avg_borough_price_1998/avg_borough_price_2018
    # Return the borough's data
    return pd.DataFrame({
        'London Borough': borough,
        'Average Price 1998': avg_borough_price_1998,
        'Average Price 2018': avg_borough_price_2018,
        'Average Price Ratio': avg_borough_price_2018/ avg_borough_price_1998
    })
# Initialize an empty dataframe for the average price ratios
average_price_ratios = pd.DataFrame()
# Loop through each London Borough
for borough in london_boroughs:
    # Concatenate current borough's data to the average_price_ratios dataframe
    average_price_ratios = pd.concat([average_price_ratios, create_price_ratio(borough)])
# Sort the average_price_ratios dataframe by Average Price Ratio in descending order
average_price_ratios.sort_values('Average Price Ratio', ascending = False, inplace=True)
# Reset the index
average_price_ratios.reset_index(drop=True, inplace=True)
# View the results
average_price_ratios
```
